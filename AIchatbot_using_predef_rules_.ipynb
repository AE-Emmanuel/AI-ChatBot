```python
# @title GEt weather Function

def get_weather(city):
    api_key = 'your key here'
    base_url = 'http://api.openweathermap.org/data/2.5/weather'
    param = {
        'q': city,
        'appid': api_key,
        'units': 'metric'
    }
    response = requests.get(base_url, params=param)
    data = response.json()
    if data['cod'] == 200:
        weather = data['weather'][0]['description']
        temperature = data['main']['temp']
        return f"The weather in {city} is currently {weather} with a temperature of {temperature}°C."
    else:
        return f"Sorry, I couldn't find the weather information for {city}."
```

```python
# @title Time Function
def get_real_time(location):
    api_key = 'your key here'
    base_url = 'http://api.timezonedb.com/v2.1/get-time-zone'
    params = {
        'key': api_key,
        'format': 'json',
        'by': 'position',
        'lat': '',  # Latitude of the location
        'lng': '',  # Longitude of the location
        'fields': 'formatted'
    }
    # Use a geocoding API to get latitude and longitude for the location
    geocoding_url = f'https://maps.googleapis.com/maps/api/geocode/json?address={location}&key=your key here'
    geocoding_response = requests.get(geocoding_url)
    geocoding_data = geocoding_response.json()
    if geocoding_data['status'] == 'OK':
        lat = geocoding_data['results'][0]['geometry']['location']['lat']
        lng = geocoding_data['results'][0]['geometry']['location']['lng']
        params['lat'] = lat
        params['lng'] = lng
        response = requests.get(base_url, params=params)
        time_data = response.json()
        current_time = time_data['formatted']
        return f"The current time in {location} is {current_time}."
    else:
        return f"Sorry, I couldn't find the location {location}."
```

```python
# @title ChatBot Main Function
def ai_chat_bot():
    print("Hello! I'm your AI chatbot - Created Using a predefined set of rules . Type exit or Bye to end the conversation.")
    while True:
        user_input = input("Me: ")
        response = chatbot.respond(user_input)
        if user_input.lower() in ['exit'] :
            print("Chatbot: Goodbye!")
            break




        if "weather" in user_input:
            # Extract city name from user input
            city = user_input.split("in")[-1].strip() if "in" in user_input else None
            while not city:
                city = input("Chatbot: Please provide your location: ").strip()
            response = get_weather(city)
        elif "time" in user_input:
            location = user_input.split("in")[-1].strip() if "in" in user_input else None
            while not location:
                location = input("Chatbot: Please provide a location: ").strip()
            response = get_real_time(location)
        else:
            response = chatbot.respond(user_input)
        print(f"Chatbot: {response}")
        if response in ["Goodbye!", "See you later!", "Adios!", "Farewell!", "Have a great day!"]:
            break

```

```python
# @title Intiate Chat Bot
if __name__ == "__main__":
    ai_chat_bot()
```
